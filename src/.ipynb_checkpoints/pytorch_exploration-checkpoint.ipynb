```python
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.__version__
```

```python
# Linear Regression

weight = 0.7
bias = 0.3

start = 0 
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(1) 
y = weight * X + bias

X[:10], y[:10], len(X), len(y)
```

```python
# Training and Test Sets
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test,  y_test = X[train_split:], y[train_split:]
```

```python
len(X_train), len(y_train), len(X_test),  len(y_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test, y_test)
```

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    plt.figure(figsize=(10,7))

    plt.scatter(train_data, train_labels, c='b', s=4, label="Training data")
    plt.scatter(test_data, test_labels, c='g', s=4, label="Testing data")

    if predictions is not None:
        plt.scatter(test_data, predictions, c='r', s=4, label="Predictions")

    plt.legend(prop={"size" : 14})
```

```python
plot_predictions()
```

```python
from torch import nn

# Build a pytorch model to build a linear regression model
class LinearRegressionModel(nn.Module):
    def __init__(self, num_feat=1):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(num_feat,
                                                requires_grad=True,
                                                dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1,
                                                requires_grad=True,
                                                dtype=torch.float))

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        if type(x) is pd.DataFrame:
            x = torch.Tensor(x.values)
        return (self.weights * x + self.bias).sum(dim=1)
```

```python
torch.manual_seed(2)

model_0 = LinearRegressionModel()

list(model_0.parameters())
```

```python
model_0.state_dict()
```

```python
weight, bias
```

```python
# Inference
with torch.inference_mode():
    y_preds =  model_0(X_test)
y_preds
```

```python
plot_predictions(predictions=y_preds)
```

```python
torch.manual_seed(1)

model_0 = LinearRegressionModel()

list(model_0.parameters())

# Set up a loss function
loss_fn = nn.L1Loss()

# Setup up optimizer
optimizer = torch.optim.SGD(params=model_0.parameters(),
                            lr=0.01)
```

```python
loss_fn
```

```python
# Training Loop
# 0. Loop through data
# 1. Forward pass
# 2. Calculate the loss
# 3. Optimizer zero grad
# 4. Loss backwards - calculates gradients of each paramter with respect to the loss
# 5. Optimizer step - adjust models parameters to improve loss 
```

```python
epochs = 200

import numpy as np
param = np.empty((0,3))

for epoch in range(epochs):
    model_0.train()

    y_pred = model_0(X_train)
    loss = loss_fn(y_pred, y_train)
    print(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    param = np.append(param,[[loss.item(), model_0.state_dict()['weights'][0], model_0.state_dict()['bias'][0]]], axis=0)
    
    model_0.eval()
```

```python
import sklearn

torch.manual_seed(1)

# a classic housing price dataset
X,y = shap.datasets.california(n_points=1000)
X_train, y_train = X.iloc[:int(len(X)*0.8)], y[:int(len(y)*0.8)]
X_test, y_test = X.iloc[int(len(X)*0.8):], y[int(len(y)*0.8):]

X100 = shap.utils.sample(X, 100) # 100 instances for use as the background distribution

# a simple linear model
#model_1 = sklearn.linear_model.LinearRegression()
model_1 = LinearRegressionModel(len(X.columns))

list(model_1.parameters())

# Set up a loss function
loss_fn = nn.L1Loss()

# Setup up optimizer
optimizer = torch.optim.SGD(params=model_1.parameters(),
                            lr=1e-7)
```

```python
epochs = 5000

import copy
import numpy as np
param = np.empty((0,1+len(model_1.weights)+len(model_1.bias)))
model_checkpoints = []

for epoch in range(epochs):
    model_1.train()

    y_pred = model_1(torch.tensor(X_train.values))
    loss = loss_fn(y_pred, torch.tensor(y_train))
    #print(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    param = np.append(param,[[loss.item(), *list(model_1.state_dict()['weights']), *list(model_1.state_dict()['bias'])]], axis=0)

    #explainer = shap.Explainer(model_1.forward, X100)
    #shap_values = explainer(X[:3])
    #shap_plots.append(shap.plots.waterfall(shap_values[0], max_display=14, show=False))

    if epoch % 100 == 0:
        print("Epoch :", epoch)
        model_checkpoints.append(copy.deepcopy(model_1))
    model_1.eval()
```

```python
for i in range(len(model_checkpoints)):
    shap_model = model_checkpoints[i]
    explainer = shap.Explainer(shap_model.forward, X100)
    shap_values = explainer(X[:3])
    waterfall_plot = shap.plots.waterfall(shap_values[0], max_display=14, show=False)
    print(waterfall_plot)
    waterfall_plot.savefig(f'wf_plot_{i}.png')
    break
```
